```python

final_df['Current_Year']=2021
```

```python
#Fiding the total no of years used for each car
final_df['No_Years']=final_df['Current_Year'] - final_df['Year']
```

```python
final_df.drop(['Year'],axis =1,inplace = True)
final_df.head()
```

```python
final_df.drop(['Current_Year'],axis =1,inplace = True)
final_df.head()
```

```python
#replace the categorical values with dummy values, drop_first=True used to avoid dummy variable trap)
final_df = pd.get_dummies(final_df,drop_first=True)
```

```python
X=final_df.iloc[:,1:]
y= final_df.iloc[:,0]
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,y)
```

```python
print(model.feature_importances_)
```

```python
#plot graph of feature importances for better visualization 
feat_importances = pd.Series(model.feature_importances_, index=X.columns) 
feat_importances.nlargest(5).plot(kind='barh') 
plt.show()
```
